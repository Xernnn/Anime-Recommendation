```python
import pandas as pd

# Read the user's anime data from a CSV file
def read_user_anime(csv_file):
    df = pd.read_csv(csv_file)
    return df
```

```python
# Read user's anime data
user_anime_file = './data.csv'
user_anime_df = read_user_anime(user_anime_file)
user_anime_df
```

```python
# Select the 5 highest-score anime
top_anime_df = user_anime_df.nlargest(5, 'Score')
top_anime_df
```

```python
# Read the anime clusters file
def read_anime_clusters(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        clusters = []
        for line in file:
            if line.startswith("Cluster #"):
                current_cluster = []
                current_cluster.append(line)
                clusters.append(current_cluster)
            elif line.startswith("\tName:"):
                anime_name = line.split("Name: ")[1].strip()
                current_cluster.append(anime_name)
    return clusters
```

```python
# Read anime clusters file
anime_clusters_file = './anime_clusters.txt'
anime_clusters = read_anime_clusters(anime_clusters_file)
anime_clusters
```

```python
anime_clusters_file = open('./anime_clusters.txt','r')
line = anime_clusters_file.readline()
print(line)
```

```python
# Find the cluster containing the user's anime
def find_cluster(user_anime_name, clusters):
    for cluster in clusters:
        for anime_name in cluster:
            if user_anime_name.lower() in anime_name.lower() or anime_name.lower() in user_anime_name.lower():
                return cluster
    return None
```

```python
# Iterate over the selected anime and find their clusters
for index, anime_row in top_anime_df.iterrows():
    user_anime_name = anime_row['Name']
    user_cluster = find_cluster(user_anime_name, anime_clusters)
    
    # Output the anime in the same cluster
    if user_cluster:
        print(f"Anime in the same cluster as '{user_anime_name}':")
        anime_names = user_cluster[:6]
        for anime_name in anime_names:
            anime_name = anime_name.split(", Score:")[0].strip()
            if (anime_name != user_anime_name): 
                print(anime_name)
        print()
    else:
        print(f"No cluster found for '{user_anime_name}'.")
        print()
```
